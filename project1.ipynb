```python
import matplotlib.pyplot as plt
import numpy as np
```

# Task2

```python
def draw_chain(n):
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    angles = np.append(angles, angles[0])
    cos = np.cos(angles)
    sine = np.sin(angles)
    plt.plot(cos, sine, zorder=1)
    plt.scatter(cos, sine, color='red', zorder=2)
    plt.show()
```

```python
for n in range(1, 6):
    draw_chain(n)
```

# Task4

```python
def draw_weighted_chain(weights):
    n = len(weights)
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    angles = np.append(angles, angles[0])
    cos = np.cos(angles)
    sine = np.sin(angles)
    for i in range(n):
        plt.plot(
            [cos[i], cos[i+1]],
            [sine[i], sine[i+1]],
            color='royalblue',
            linewidth=weights[i],
            zorder=1
        )
    plt.scatter(cos, sine, color='red', s=100, zorder=2)
    plt.show()
```

```python
weights = [1, 2, 6, 4, 5, 9, 2]
draw_weighted_chain(weights)
```

# Task6

```python
# adjacency matrix
n = len(weights)
W = np.zeros((n, n))

for i in range(n):
    W[i, (i+1)%n] = weights[i]
    W[(i+1)%n, i] = weights[i]
W
```

```python
degrees = np.sum(W, axis=1)
degrees
```
