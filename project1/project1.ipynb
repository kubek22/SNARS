```python
import matplotlib.pyplot as plt
import numpy as np
```

# Task1

```python
def draw_square_lattice(width, height):
    xs = list(range(0, width)) * height
    ys = [i//width for i in range(height*width)]
    x_cords = []
    y_cords = []
    for i in range(width * height-1):
        for j in range(i+1, width * height):
            if abs(xs[i] - xs[j]) + abs(ys[i] - ys[j]) == 1:
                x_cords.append((xs[i], xs[j]))
                y_cords.append((ys[i], ys[j]))
    
    plt.scatter(xs, ys, c='red', s=100, zorder=2)
    for i in range(len(x_cords)):
        plt.plot(x_cords[i], y_cords[i], 'b-', zorder=1)
        
   

    plt.show()
```

```python
draw_square_lattice(7,5)
```

# Task2

```python
def draw_chain(n):
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    angles = np.append(angles, angles[0])
    cos = np.cos(angles)
    sine = np.sin(angles)
    plt.plot(cos, sine, zorder=1)
    plt.scatter(cos, sine, color='red', zorder=2)
    plt.show()
```

```python
for n in range(1, 6):
    draw_chain(n)
```

# Task3

```python
def draw_complete(n):
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    angles = np.append(angles, angles[0])
    xs = np.cos(angles)
    ys = np.sin(angles)
    x_cords = []
    y_cords = []
    for i in range(n-1):
        for j in range(i+1, n):
            x_cords.append((xs[i], xs[j]))
            y_cords.append((ys[i], ys[j]))
    
    plt.scatter(xs, ys, c='red', s=100, zorder=2)
    for i in range(len(x_cords)):
        plt.plot(x_cords[i], y_cords[i], 'b-', zorder=1)
        
   

    plt.show()
```

```python
draw_complete(10)
```

# Task4

```python
def draw_weighted_chain(weights):
    n = len(weights)
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    angles = np.append(angles, angles[0])
    cos = np.cos(angles)
    sine = np.sin(angles)
    for i in range(n):
        plt.plot(
            [cos[i], cos[i+1]],
            [sine[i], sine[i+1]],
            color='royalblue',
            linewidth=weights[i],
            zorder=1
        )
    plt.scatter(cos, sine, color='red', s=100, zorder=2)
    plt.show()
```

```python
weights = [1, 2, 6, 4, 5, 9, 2]
draw_weighted_chain(weights)
```

# Task5

```python
def draw_directed(v,e):
    angles = np.linspace(0, 2 * np.pi, len(v), endpoint=False)
    angles = np.append(angles, angles[0])
    xs = np.cos(angles)
    ys = np.sin(angles)
    
    for v1,v2 in e:
        plt.arrow(xs[v.index(v1)], ys[v.index(v1)], xs[v.index(v2)] - xs[v.index(v1)], ys[v.index(v2)] - ys[v.index(v1)], width = 0.02, zorder=1, length_includes_head=True)
    
    plt.scatter(xs, ys, c='red', s=100, zorder=2)
        
   

    plt.show()
```

```python
v = [2,5,7,8,9, 'c']
e =[(2,5), (5,2), (7,8), (8,2), (2,7), (9, 5), ('c', 7)]
draw_directed(v,e)
```
